```python
id_vars = ['Country', 'Country Code', 'Series', 'Series Code']
melted_data = filtered_data.melt(id_vars=id_vars, var_name='Year', value_name='Value')
melted_data
```

The dataset contain information for each quarter in the year, each month in the year and the closing value for the year which is overall average of the values throughput the year. Only the overall average is needed so that will be extracted and the rest will be removed.

Since the average for 2025 is not yet calculated, that will be done manually

```python
import re

def extract_year(year_str):
    match = re.search(r'\d{4}', year_str)
    if match:
        return match.group(0)
    return None

melted_data['Extracted_Year'] = melted_data['Year'].apply(extract_year)

#filter the years
filtered_years_data = melted_data[
    (melted_data['Extracted_Year'].astype(int) >= 2025) |
    (melted_data.apply(lambda row: row['Year'] == f"{row['Extracted_Year']} [{row['Extracted_Year']}]", axis=1))
]

filtered_years_data = filtered_years_data.drop(columns=['Extracted_Year'])

#view 10 samples
filtered_years_data.sample(10)
```

The next step now is to compute for the overall average for 2025 with the available data. The dataset was last updated on 10/24/2025

```python
import numpy as np
import warnings
warnings.filterwarnings('ignore')
#filter data for 2025
data_2025 = filtered_years_data[filtered_years_data['Year'].str.contains('2025')]

#convert the value to number
data_2025['Value'] = pd.to_numeric(data_2025['Value'], errors='coerce')
data_2025['Value'] = data_2025['Value'].replace('..', np.nan)
data_2025['Value'] = pd.to_numeric(data_2025['Value'])

#calculate the average for each group
averaged_2025_data = data_2025.groupby(['Country', 'Country Code', 'Series', 'Series Code'])['Value'].mean().reset_index()
averaged_2025_data['Year'] = '2025 [2025]'

#remove the original 2025 data
filtered_years_data_before_2025 = filtered_years_data[~filtered_years_data['Year'].str.contains('2025')]

#concatenate the data before 2025 with the averaged 2025 data
final_data = pd.concat([filtered_years_data_before_2025, averaged_2025_data], ignore_index=True)

final_data.sample(15)
```

### Stationarity Tests (ADF and KPSS)

Before running any time-series analysis, it is important to check whether each economic series is stationary. A stationary series is one whose statistical properties such as mean and variance do not change over time. Many econometric models and causality tests require stationarity to work properly.

Four variables will be tested for each country:

* stock_usd_ln – stock index in USD (log)

* stock_lcu_ln – stock index in local currency (log)

* exch_ln – exchange rate (log)

* index_gap – deviation of stock index from long-run trend

To determine whether these series are stationary or need differencing, two complementary tests will be used:

* Augmented Dickey–Fuller (ADF) Test

  ADF checks whether a series has a unit root (i.e., whether it is non-stationary).

  Null hypothesis: the series is non-stationary.

  If the p-value < 0.05, we reject the null and the series is considered stationary.

* KPSS Test

  KPSS works in the opposite direction of the ADF test.

  Null hypothesis: the series is stationary.

  If the p-value > 0.05, we fail to reject the null and the series is stationary.


ADF and KPSS complement each other.
Using both helps avoid wrong conclusions that could come from relying on only one test.

```python
#create the function for adf test
def adf_test(series, name="series", regress='c'):
    series = series.dropna()
    if len(series) < 10 or series.std() == 0:
        return {'adf_stat': np.nan, 'pvalue': np.nan, 'nobs': len(series)}
    res = adfuller(series, regression=regress, autolag='AIC')
    return {'adf_stat': res[0], 'pvalue': res[1], 'nobs': res[3]}

#create the function for kpss test
def kpss_test(series, name="series", regression='c'):
    series = series.dropna()
    if len(series) < 10 or series.std() == 0:
        return {'kpss_stat': np.nan, 'pvalue': np.nan, 'nobs': len(series)}
    try:
        res = kpss(series, regression=regression, nlags="auto")
        return {'kpss_stat': res[0], 'pvalue': res[1], 'nobs': len(series)}
    except Exception as e:
        return {'kpss_stat': np.nan, 'pvalue': np.nan, 'nobs': len(series), 'error': str(e)}
```

```python
import warnings
warnings.filterwarnings('ignore')
#run the tests for all the countries and get the summary
countries = df.index.get_level_values(0).unique()

stationarity_results = []
for country in countries:
    sub = df.xs(country)
    for col in ['stock_usd_ln', 'stock_lcu_ln', 'exch_ln', 'index_gap']:
        if col not in sub:
            continue
        a = adf_test(sub[col].dropna(), name=f"{country}-{col}")
        k = kpss_test(sub[col].dropna(), name=f"{country}-{col}")
        stationarity_results.append({
            'country': country, 'series': col,
            'adf_stat': a['adf_stat'], 'adf_p': a['pvalue'],
            'kpss_stat': k.get('kpss_stat', np.nan), 'kpss_p': k.get('pvalue', np.nan),
            'nobs': a['nobs']
        })

stationarity_df = pd.DataFrame(stationarity_results)
#display the summary
stationarity_df['adf_stationary'] = stationarity_df['adf_p'] < 0.05
stationarity_df['kpss_stationary'] = stationarity_df['kpss_p'] > 0.05
stationarity_summary = stationarity_df.groupby('series')[['adf_stationary','kpss_stationary']].mean()
print(stationarity_summary)
```

The results indicate that most series show low proportions of stationarity under both ADF and KPSS tests, meaning the variables are generally non-stationary in levels across countries. Exchange rates, index gaps, and both stock index measures fail to meet stationarity requirements for the majority of countries, implying the presence of trends or persistent shocks

### Cointegration Checks
After checking stationarity, the next step is to examine whether a long-run relationship exists between the exchange rate and stock market indices for each country. Even if two series are individually non-stationary, they may still move together in a stable long-term pattern.

```python
cointegrated = []
for country in countries:
    sub = df.xs(country)
    if sub[['stock_usd_ln','exch_ln']].dropna().shape[0] < 10:
        continue
    try:
        tstat, pvalue, _ = coint(sub['stock_usd_ln'].dropna(), sub['exch_ln'].dropna())
        cointegrated.append({'country': country, 'tstat': tstat, 'pvalue': pvalue})
    except Exception as e:
        cointegrated.append({'country': country, 'tstat': np.nan, 'pvalue': np.nan, 'error': str(e)})

coin_df = pd.DataFrame(cointegrated).sort_values('pvalue')
coin_df.head(10)
```

These top cointegration results show the countries with the strongest statistical evidence that their stock index (USD-based) and exchange rate move together in a long-run equilibrium relationship. Extremely large negative test statistics and p-values of essentially zero (as seen for Pakistan and Uganda) indicate very strong cointegration, meaning deviations between stock prices and exchange rates tend to correct over time.

Countries like Ukraine, the Slovak Republic, Brazil, and Sri Lanka also show statistically significant cointegration, though with weaker evidence compared to the top cases. Overall, these results suggest that in several countries, stock markets and exchange rates are tightly linked in the long run, despite short-term fluctuations.

### Cross-Sectional Dependence Test
The cross-sectional dependence test (Pesaran CD test) checks whether countries in the panel dataset are influenced by common global factors. After running a pooled regression, the correlations in the residuals across countries are examined. If these residuals move together, it means that countries share underlying shocks such as global financial trends, commodity price changes, or regional crises. Detecting this dependence is important because ignoring it can make regression results unreliable or misleading.

```python
panel_df = df.reset_index().dropna(subset=['stock_usd_ln', 'exch_ln', 'gdp_usd', 'unemp_pct'] , how='any')
macro_vars = []
if 'gdp_usd' in panel_df.columns: macro_vars.append('gdp_usd')
if 'unemp_pct' in panel_df.columns: macro_vars.append('unemp_pct')
if 'exch_ln' in panel_df.columns: macro_vars.append('exch_ln')
cols_needed = ['Country','Year','stock_usd_ln'] + macro_vars
panel_df = panel_df[cols_needed].dropna()

#simple pooled ols
panel_df = panel_df.set_index(['Country','Year'])
y = panel_df['stock_usd_ln']
X = panel_df[macro_vars]
X = sm.add_constant(X)

#fit pooled ols via linearmodels
from linearmodels.panel import PooledOLS
pooled = PooledOLS(y, X)
pooled_res = pooled.fit(cov_type='robust')
print(pooled_res.summary)
```

The Pesaran CD statistic therefore helps confirm whether countries behave independently or whether their stock–exchange rate dynamics are interconnected at a global level.

```python
#compute residuals by country-time and then pivot to residual matrix
residuals = pooled_res.resids.reset_index()
res_pivot = residuals.pivot(index='Country', columns='Year', values='residual')
res_pivot = res_pivot.dropna(axis=0, how='any')
r = res_pivot.T.corr().values
n = r.shape[0]
#pesaran CD statistic
upper_idx = np.triu_indices(n, k=1)
pairwise = r[upper_idx]
cd_stat = np.sqrt(2/(n*(n-1))) * pairwise.sum() * np.sqrt(res_pivot.shape[1])
print("Approx Pesaran CD stat (heuristic):", cd_stat)
```

The Pesaran CD statistic of 36.59 shows very strong cross-sectional dependence, meaning countries in the data are highly interconnected and influenced by common shocks rather than behaving independently.

### Panel-Fixed Effects Regression

The panel fixed-effects regression is the main statistical test used to check whether exchange rate movements truly influence stock market performance after accounting for each country’s unique characteristics. The fixed effects model controls for all country specific factors that do not change over time such as economic structure, governance quality, or long-term monetary policy so the analysis focuses only on year-to-year changes within each country. By including macroeconomic controls (GDP and unemployment) and clustering the standard errors, the model provides a more reliable estimate of how strongly exchange rates affect stock indices. This regression forms the core evidence for determining whether exchange rate fluctuations systematically drive stock market behavior across countries.

```python
panel = df[['stock_usd_ln','exch_ln','gdp_usd','unemp_pct']].copy()
panel = panel.dropna()
panel['gdp_usd_ln'] = np.log(panel['gdp_usd'].replace(0, np.nan))
exog_vars = []
if 'exch_ln' in panel.columns: exog_vars.append('exch_ln')
if 'gdp_usd_ln' in panel.columns: exog_vars.append('gdp_usd_ln')
if 'unemp_pct' in panel.columns: exog_vars.append('unemp_pct')

panel = panel.dropna(subset=['stock_usd_ln']+exog_vars)
y = panel['stock_usd_ln']
X = sm.add_constant(panel[exog_vars])
panel = panel.reset_index().set_index(['Country','Year'])

from linearmodels.panel import PanelOLS
fe_mod = PanelOLS(panel['stock_usd_ln'], sm.add_constant(panel[exog_vars]), entity_effects=True)
fe_res = fe_mod.fit(cov_type='clustered', cluster_entity=True)
print(fe_res.summary)
```

The results show that the exchange rate is the strongest driver. When the exchange rate moves in a way that makes the local currency weaker against the dollar, the stock index appears lower once converted to USD, even if nothing else changes. Unemployment also shows a small negative effect, meaning higher unemployment is associated with slightly lower stock values in USD terms.

The GDP does not show a meaningful link in this model. The test also confirms that each country has unique characteristics that must be controlled for.

Overall, movements in the exchange rate are the main factor behind how strong or weak a country’s stock market looks when expressed in USD.

### Granger causality test

This section applies per-country Granger causality tests to determine whether exchange-rate movements help predict stock-market performance or whether stock-market movements help predict exchange rates. Significance is assessed using F-test p-values at the chosen lag, and results indicate which countries exhibit predictive relationships from exchange rates to stocks, from stocks to exchange rates, or both.

```python
from statsmodels.tsa.stattools import grangercausalitytests

maxlag = 3

granger_results = []
for country in countries:
    sub = df.xs(country).dropna(subset=['stock_usd_ln','exch_ln'])
    if len(sub) < (maxlag + 5):
        continue
    data = sub[['stock_usd_ln','exch_ln']]
    try:
        test_exch_to_stock = grangercausalitytests(data[['stock_usd_ln','exch_ln']], maxlag=maxlag, verbose=False)
        p_exch_to_stock = test_exch_to_stock[maxlag][0]['ssr_ftest'][1]
    except Exception as e:
        p_exch_to_stock = np.nan
    try:
        test_stock_to_exch = grangercausalitytests(data[['exch_ln','stock_usd_ln']], maxlag=maxlag, verbose=False)
        p_stock_to_exch = test_stock_to_exch[maxlag][0]['ssr_ftest'][1]
    except Exception as e:
        p_stock_to_exch = np.nan

    granger_results.append({
        'country': country,
        'p_exch_to_stock': p_exch_to_stock,
        'p_stock_to_exch': p_stock_to_exch,
        'nobs': len(sub)
    })

granger_df = pd.DataFrame(granger_results).dropna(subset=['p_exch_to_stock','p_stock_to_exch'])
#Flag significance at 0.05
granger_df['exch_causes_stock'] = granger_df['p_exch_to_stock'] < 0.05
granger_df['stock_causes_exch'] = granger_df['p_stock_to_exch'] < 0.05
granger_df.sort_values('nobs', ascending=False).head(10)
```

```python
significant_granger_causality = granger_df[(granger_df['exch_causes_stock'] == True) | (granger_df['stock_causes_exch'] == True)]

print(f"Number of countries with significant Granger causality (either direction): {len(significant_granger_causality)}")

if not significant_granger_causality.empty:
    print("Countries exhibiting significant Granger causality:")
    display(significant_granger_causality)
else:
    print("No countries show significant Granger causality in either direction.")
```

### Analzying these countries with significant Granger causality

```python
#countries identified
selected_countries = ["India", "Malaysia", "Qatar", "Saudi Arabia", "Thailand"]

#containers for summary results
stationarity_results = []
cointegration_results = []
regression_results = []
ecm_results = []

#loop through each country
for country in selected_countries:
    print(f"Processing: {country}")
    data = df.loc[country].copy()
    data = data.sort_values("Year")

    #1. STATIONARITY TESTS
    def adf_test(series):
        return adfuller(series.dropna(), autolag="AIC")[1]

    def kpss_test(series):
        return kpss(series.dropna(), nlags="auto")[1]

    adf_exch = adf_test(data["exch_ln"])
    kpss_exch = kpss_test(data["exch_ln"])
    adf_stock = adf_test(data["stock_usd_ln"])
    kpss_stock = kpss_test(data["stock_usd_ln"])

    stationarity_results.append([country, adf_exch, kpss_exch, adf_stock, kpss_stock])

    #2. COINTEGRATION TEST
    coint_pvalue = coint(data["stock_usd_ln"], data["exch_ln"])[1]
    cointegration_results.append([country, coint_pvalue])

    #3. OLS REGRESSION
    if 'gdp_usd_ln' not in data.columns:
        data['gdp_usd_ln'] = np.log(data['gdp_usd'].replace(0, np.nan))

    reg_df = data[['stock_usd_ln','exch_ln','gdp_usd_ln','unemp_pct']].dropna()

    y = reg_df['stock_usd_ln']
    X = sm.add_constant(reg_df[['exch_ln','gdp_usd_ln','unemp_pct']])

    ols_model = sm.OLS(y, X).fit()
    regression_results.append([country, ols_model.params, ols_model.pvalues, ols_model.rsquared])

    print("\nOLS Regression Summary:")
    print(ols_model.summary())

    #4. ECM MODEL (ONLY IF COINTEGRATION EXISTS)
    if coint_pvalue < 0.05:
        print(f"\nECM is appropriate: Cointegration detected (p={coint_pvalue:.4f})")

        #carry out long-run regression
        long_run = sm.OLS(y, X).fit()
        data["ECT"] = long_run.resid

        #build ECM
        data["d_stock"] = data["stock_usd_ln"].diff()
        data["d_exch"] = data["exch_ln"].diff()
        data["d_gdp"] = data["gdp_usd_ln"].diff()
        data["d_unemp"] = data["unemp_pct"].diff()
        data["ECT_lag"] = data["ECT"].shift(1)

        ecm_df = data[['d_stock','d_exch','d_gdp','d_unemp','ECT_lag']].dropna()
        y_ecm = ecm_df['d_stock']
        X_ecm = sm.add_constant(ecm_df[['d_exch','d_gdp','d_unemp','ECT_lag']])

        ecm_model = sm.OLS(y_ecm, X_ecm).fit()
        ecm_results.append([country, ecm_model.params, ecm_model.pvalues, ecm_model.rsquared])

        print("\nECM Summary:")
        print(ecm_model.summary())

    else:
        print(f"No ECM for {country} (no cointegration, p={coint_pvalue:.4f})")

```

```python
#convert the results to dataframe
stationarity_df = pd.DataFrame(stationarity_results,
    columns=["Country","ADF_exch","KPSS_exch","ADF_stock","KPSS_stock"])

cointegration_df = pd.DataFrame(cointegration_results,
    columns=["Country","Cointegration_pvalue"])

regression_df = pd.DataFrame(regression_results,
    columns=["Country","Coefficients","Pvalues","R2"])

ecm_df = pd.DataFrame(ecm_results,
    columns=["Country","ECM_Coeff","ECM_Pvalues","ECM_R2"])

#print the final summary tables
print("\nSTATIONARITY RESULTS")
display(stationarity_df)

print("\nCOINTEGRATION RESULT")
display(cointegration_df)

print("\nREGRESSION RESULTS (OLS)")
display(regression_df)

print("\nECM RESULTS (only cointegrated countries)")
display(ecm_df)
```

```python
def plot_regression_generic(
        country,
        data,
        ols_model_to_plot,
        stock_col,
        exch_col,
        stock_label="Stock Market",
        exch_label="Exchange Rate"
    ):

    #data for the scatter plot
    plot_df = data[[stock_col, exch_col]].dropna()
    X_predict_df = pd.DataFrame(columns=ols_model_to_plot.params.index, index=plot_df.index)
    X_predict_df['const'] = 1.0
    X_predict_df[exch_col] = plot_df[exch_col]
    other_predictors = [col for col in ols_model_to_plot.params.index if col not in ['const', exch_col]]
    for col in other_predictors:
        if col in data.columns:
            X_predict_df[col] = data[col].mean()
        elif col == 'gdp_usd_ln' and 'gdp_usd' in data.columns:
            temp_gdp_ln = np.log(data['gdp_usd'].replace(0, np.nan))
            X_predict_df[col] = temp_gdp_ln.mean()
        elif col == 'unemp_pct' and 'unemp_pct' in data.columns:
            X_predict_df[col] = data['unemp_pct'].mean()
        else:
            print(f"Warning: Predictor '{col}' not found in data for setting mean. Setting to 0.")
            X_predict_df[col] = 0.0
    X_predict_df = X_predict_df[ols_model_to_plot.params.index]
    y_pred = ols_model_to_plot.predict(X_predict_df)

    #create the scatter plot
    sns.scatterplot(
        x=plot_df[exch_col],
        y=plot_df[stock_col],
        alpha=0.6,
        label=f"{country}: Data"
    )

    #add the regression line
    plt.plot(
        plot_df[exch_col],
        y_pred,
        color="red",
        label=f"{country}: Regression Line (other vars at mean)"
    )
    plt.title(f"{country} — Regression of {stock_label} on {exch_label}")
    plt.xlabel(exch_label)
    plt.ylabel(stock_label)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

### Clustering

The final aspect of this analysis is to carry out clustering using K-means to group the countries data into optimal number of clusters that will be determined using elbow method

```python
#import basic required libraries
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np
```

```python
cluster_df = pd.DataFrame()
#correctly get unique countries from the multiindex
cluster_df['Country'] = df.index.get_level_values('Country').unique()

#lists to store features per country
corr_list = []
corr_lcu_list = []
exch_vol_list = []
stock_usd_vol_list = []
stock_lcu_vol_list = []
gdp_mean_list = []
unemp_mean_list = []
stock_usd_mean_list = []
stock_lcu_mean_list = []
exch_mean_list = []

for country in cluster_df['Country']:
    sub = df.loc[country].copy()
    sub = sub.sort_values("Year")

    #usd stock correlation with exchange rate
    corr_usd = sub[['exch_ln', 'stock_usd_ln']].corr().iloc[0,1]
    corr_list.append(corr_usd)

    #lcu stock correlation with exchange rate
    corr_lcu = sub[['exch_ln', 'stock_lcu_ln']].corr().iloc[0,1]
    corr_lcu_list.append(corr_lcu)

    #volatilities
    exch_vol_list.append(sub['exch_ln'].std())
    stock_usd_vol_list.append(sub['stock_usd_ln'].std())
    stock_lcu_vol_list.append(sub['stock_lcu_ln'].std())
    if 'gdp_usd_ln' not in sub.columns:
        sub['gdp_usd_ln'] = np.log(sub['gdp_usd'].replace(0, np.nan))

    gdp_mean_list.append(sub['gdp_usd_ln'].mean())
    unemp_mean_list.append(sub['unemp_pct'].mean())

    #stock price levels
    stock_usd_mean_list.append(sub['stock_usd_ln'].mean())
    stock_lcu_mean_list.append(sub['stock_lcu_ln'].mean())
    exch_mean_list.append(sub['exch_ln'].mean())

#assign columns
cluster_df['corr_exch_stock_usd'] = corr_list
cluster_df['corr_exch_stock_lcu'] = corr_lcu_list
cluster_df['exch_vol'] = exch_vol_list
cluster_df['stock_usd_vol'] = stock_usd_vol_list
cluster_df['stock_lcu_vol'] = stock_lcu_vol_list
cluster_df['gdp_mean_ln'] = gdp_mean_list
cluster_df['unemp_mean'] = unemp_mean_list
cluster_df['stock_usd_mean'] = stock_usd_mean_list
cluster_df['stock_lcu_mean'] = stock_lcu_mean_list
cluster_df['exch_mean'] = exch_mean_list
#add granger results data
granger_df.rename(columns={'country': 'Country'}, inplace=True)
cluster_df = cluster_df.merge(granger_df[['Country', 'p_exch_to_stock', 'p_stock_to_exch']], on='Country', how='left')
```

```python
#scale the data
cluster_data = cluster_df.drop(columns=['Country']).fillna(0)
scaler = StandardScaler()
scaled = scaler.fit_transform(cluster_data)
```

```python
#using elbow method to determine optimal number of clusters
inertias = []
K_range = range(2, 10)

for k in K_range:
    km = KMeans(n_clusters=k, n_init=20, random_state=42)
    km.fit(scaled)
    inertias.append(km.inertia_)
plt.plot(K_range, inertias)
plt.xlabel("K")
plt.ylabel("Inertia")
plt.title("Elbow Plot for Determining K")
plt.show()
```

```python
k_opt = 3
kmeans = KMeans(n_clusters=k_opt, n_init=20, random_state=42)
cluster_df['Cluster'] = kmeans.fit_predict(scaled)

cluster_df.head()
```

```python
#visualize the average of scaled features after clustering
sns.heatmap(
    cluster_df.drop(columns=['Country']).groupby("Cluster").mean(),
    annot=True,
    cmap="viridis",
    cbar = False
)

plt.title("Cluster Feature Heatmap")
plt.show()
```

```python
#exchange rate and stock indices by clusters
clusters = cluster_df["Cluster"].unique()

for c in clusters:
    subset = cluster_df[cluster_df["Cluster"] == c]
    plt.scatter(
        subset["exch_mean"],
        subset["stock_usd_mean"],
        s=80,
        alpha=0.7,
        label=f"Cluster {c}"
    )

plt.xlabel("Exchange Rate (log)")
plt.ylabel("Stock Index USD (log)")
plt.title("Clusters of Countries: Exchange Rate vs Stock (USD)")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(scaled, cluster_df['Cluster'])
print(f"The average silhouette score is : {silhouette_avg}")
```
